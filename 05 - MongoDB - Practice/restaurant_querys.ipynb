```python
from mongodb_connection import MongoDB_Connection  # Made by me
from bson import json_util
```

```python
collection = MongoDB_Connection("sample_restaurants", "restaurants").connect()
```

```python
def response_printer(response_cursor, limit=5):
    response = list(response_cursor)
    response_len = len(response)

    response_printable_len = min(response_len, limit)
    first_or_only = "only" if response_len <= limit else "first"

    print(
        f"Showing the {first_or_only} {response_printable_len} restaurants ({response_len} in total):"
    )

    for restaurant in response[:(response_printable_len)]:
        print("=" * 60)
        print(json_util.dumps(restaurant, indent=4))

    if response_len > limit:
        print("=" * 30)
        print(f"{response_len-limit} results left.")
```

```python
# 1. Write a MongoDB query to display all the documents in the collection restaurants for all the documents in the collection restaurant

response = collection.find(filter={}, projection={})
response_printer(response)
```

```python
# 2. Write a MongoDB query to display the fields restaurant_id, name, borough and cuisine

response = collection.find(
    filter={},
    projection={"restaurant_id": 1, "name": 1, "borough": 1, "cuisine": 1},
)
response_printer(response)
```

```python
# 3. Write a MongoDB query to display the fields restaurant_id, name, borough and zip code, but exclude the field _id for all the documents in the collection restaurant.

response = collection.find(
    filter={},
    projection={
        "restaurant_id": 1,
        "name": 1,
        "borough": 1,
        "address": {"zipcode": 1},
        "_id": 0,
    },
)
response_printer(response)
```

```python
# 4. Write a MongoDB query to display all the restaurant which is in the borough Bronx

response = collection.find(
    filter={"borough": "Bronx"},
    projection={"name": 1, "borough": 1, "restaurant_id": 1},
)
response_printer(response)
```

```python
# 5. Write a MongoDB query to display the first 5 restaurants which are in the borough Bronx.

response = collection.find(filter={}, projection={"name": 1, "_id": 0}, limit=5)
response_printer(response)
```

```python
# 6. Write a MongoDB query to find the restaurants who achieved a score more than 90.

# Utiliza $elemMatch para buscar dentro de un array (grades) y filtrar documentos que contengan al menos una calificación que cumpla con el criterio de puntuación mayor a 90.
response = collection.find(
    filter={
        "grades": {
            "$elemMatch": {
                "score": {"$gt": 90}  # Buscar calificaciones con un score mayor a 90
            }
        }
    },
    projection={"name": 1, "grades": 1, "restaurant_id": 1, "_id": 0},
)
response_printer(response)
```

```python
# 7. Write a MongoDB query to find the restaurants that achieved a score, more than 80 but less than 100

response = collection.find(
    filter={
        "grades": {
            "$elemMatch": {
                "score": {"$gt": 80, "$lt": 100}  # mayor a 80 pero menor a 100
            }
        }
    },
    projection={"name": 1, "grades": 1, "restaurant_id": 1, "_id": 0},
)
response_printer(response)
```

```python
# 8. Write a MongoDB query to find the restaurants that do not prepare any cuisine of 'American' and their grade score more than 70 and latitude less than -65.754168

response = collection.find(
    filter={
        "cuisine": {"$ne": "American"},  # No deben ser de cocina Americana
        "grades": {
            "$elemMatch": {
                "score": {
                    "$gt": 70
                }  # Deben tener al menos una calificación con puntuación mayor a 70
            }
        },
        "address.coord.1": {
            "$lt": -65.754168
        },  # Latitud (segundo elemento del array) menor a -65.754168
    },
    projection={"name": 1, "cuisine": 1, "grades": 1, "address": 1, "_id": 0},
)
response_printer(response)
```

```python
# 9. Write a MongoDB query to find the restaurants which do not prepare any cuisine of 'American' and achieved a score more than 70 and located in the longitude less than -65.754168.

response = collection.find(
    filter={
        "cuisine": {"$ne": "American"},  # No deben ser de cocina Americana
        "grades": {
            "$elemMatch": {
                "score": {
                    "$gt": 70
                }  # Deben tener al menos una calificación con puntuación mayor a 70
            }
        },
        "address.coord.0": {
            "$lt": -65.754168
        },  # Longitud (primer elemento del array) menor a -65.754168
    },
    projection={"name": 1, "cuisine": 1, "grades": 1, "address": 1, "_id": 0},
)
response_printer(response)
```

```python
# 10. Write a MongoDB query to find the restaurant Id, name, borough and cuisine for those restaurants which contain 'Wil' as first three letters for its name.

response = collection.find(
    filter={"name": {"$regex": "^Wil"}},
    projection={"name": 1, "borough": 1, "cuisine": 1, "restaurant_id": 1},
)
response_printer(response)
```

```python
# 11. Write a MongoDB query to find the restaurant name, borough, longitude and latitude and cuisine for those restaurants which contain 'Mad' as first three letters of its name.

response = collection.find(
    filter={"name": {"$regex": "^Mad"}},
    projection={"name": 1, "borough": 1, "address.coord": 1, "cuisine": 1, "_id": 0},
)
response_printer(response)
```
